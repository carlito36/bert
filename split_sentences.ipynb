```python
custom_tweet = list(df['sentence'])
custom_tweet
```

```python
from transformers import pipeline
classifier = pipeline("text-classification", model="matthewburke/korean_sentiment")
#custom_tweet = ["영화 재밌다.", "영화 재미없다."]
preds = classifier(custom_tweet, return_all_scores=True)
#is_positive = preds[0][1]['score'] > 0.5
```

```python
score = []
for i in range(len(preds)):
    if preds[i][0]['score'] > 0.5:
        score.append(preds[i][0]['score'])
    else :
        score.append(preds[i][1]['score'])
    
#score
```

```python
sent = []
for i in range(len(preds)):
    if preds[i][0]['score'] > 0.5:
        sent.append(preds[i][0]['label'])
    else :
        sent.append(preds[i][1]['label'])
#label
```

```python
pred = pd.DataFrame(zip(sent, score), columns=['sent', 'score'])
#pred[pred['sent']=='LABEL_0']
```

```python
pred.loc[(pred['sent'] == 'LABEL_0'), 'score'] = 1 - pred.loc[(pred['sent'] == 'LABEL_0'), 'score']
pred
```
